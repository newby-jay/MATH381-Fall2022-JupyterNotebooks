1. Initialize x₀
2. Calculate r₀ = Ax₀ − b
3. Assign p₀ = −r₀
4. For k = 0, 1, 2, …:  GUARANTEES TO CONVERGE UNDER #rows iterations
    * calculate αₖ = -rₖ'pₖ / pₖ'Apₖ
    * update xₖ₊₁ = xₖ + αₖpₖ
    * calculate rₖ₊₁ = rₖ+αₖApₖ
    * calculate βₖ₊₁ = rₖ₊₁'Apₖ / pₖ'Apₖ
    * update pₖ₊₁ = -rₖ₊₁ + βₖ₊₁pₖ

```python
# np.random.seed(0)
A = make_spd_matrix(10, random_state=0)
x_s = np.random.random(10)
b = np.dot(A,x_s)
print(x_s)

def CG(A, b, x0, tol=1e-16):
  xk = x0
  rk = np.dot(A,xk) - b
  pk = -rk
  rk_norm = np.linalg.norm(rk)

  num_iter = 0
  curve_x = [xk]
  while rk_norm > tol:
    apk = np.dot(A,pk)
    rkpk = -np.dot(rk,pk)
    alpha = rkpk / np.dot(pk,apk)

    xk = xk + alpha * pk
    rk = rk + alpha * apk

    beta = np.dot(rk,rk)/rkpk
    pk = -rk + beta*pk

    num_iter = num_iter + 1
    curve_x.append(xk)
    rk_norm = np.linalg.norm(rk)
    print('It: {} \t x = {}\t res = {:.4f}'.format(num_iter, xk, rk_norm))

    
```
